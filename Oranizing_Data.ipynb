```python
%cd ..
root_dir = "/content/small_dog_cat_dataset"
 
list_cats_test = []
list_dogs_test = []
list_cats_train = []
list_dogs_train = []
for sub_dir in os.listdir(root_dir): # listdir
  sub_dir_path = os.path.join(root_dir, sub_dir) # Đưa ra đường dẫn
  # print(sub_dir_path)
  for class_dir in os.listdir(sub_dir_path):
    class_dir_path = os.path.join(sub_dir_path, class_dir)
    # print(class_dir_path)
    for image_name in os.listdir(class_dir_path):
      image_path = os.path.join(class_dir_path, image_name)
      # read image
      image = cv2.imread(image_path)
      # show image
      # cv2_imshow(image)

      if class_dir == "cats" and sub_dir == "test":
        list_cats_test.append(image_path)
      elif class_dir == "dogs" and sub_dir == "test":
        list_dogs_test.append(image_path)
      elif class_dir == "cats" and sub_dir == "train":
        list_cats_train.append(image_path)
      elif class_dir == "dogs" and sub_dir == "train":
        list_dogs_train.append(image_path)

print(list_cats_test[:5])
print(list_dogs_test[:5])
print(list_cats_train[:5])
print(list_dogs_train[:5])
```

```python
# Hàm glob giúp truy cập đường dẫn nhanh hơn
list_cats_test = glob.glob("/content/small_dog_cat_dataset/test/cats/*.jpg")
list_dogs_test = glob.glob("/content/small_dog_cat_dataset/test/dogs/*.jpg")
list_cats_train = glob.glob("/content/small_dog_cat_dataset/train/cats/*.jpg")
list_dogs_train = glob.glob("/content/small_dog_cat_dataset/train/dogs/*.jpg")

print(list_cats_test[:5])
print(list_dogs_test[:5])
print(list_cats_train[:5])
print(list_dogs_train[:5])
```

**Lý do glob nhanh hơn os:**
os phải duyệt qua tất cả các file bên trong để đưa ra được đường dẫn đó. Còn với glob không cần phải đọc qua file đó mà chỉ cần đưa thẳng tên đó (duyệt bên ngoài chứ không vào file đọc).

```python
import matplotlib.pyplot as plt
import cv2
import glob

list_image_path = glob.glob("small_dog_cat_dataset/test/cats/*.jpg")
list_image = []

for i in range(25):
  image_path = list_image_path[i]
  image = cv2.imread(image_path)
  rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

  # resize image
  rgb_image = cv2.resize(rgb_image, (50, 50))

  list_image.append(rgb_image)

fig = plt.figure(figsize=(10, 10))
# define width and height
columns = 5
rows = 5

for i in range(1, columns*rows + 1):
  fig.add_subplot(rows, columns, i)
  plt.axis('on') # tắt hiển thị 2 trục
  plt.imshow(list_image[i-1])
```
