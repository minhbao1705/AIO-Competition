```python
import matplotlib.pyplot as plt
import cv2
import glob

list_image_path = glob.glob("small_dog_cat_dataset/test/cats/*.jpg")
list_image = []

for i in range(25):
  image_path = list_image_path[i]
  image = cv2.imread(image_path)
  rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

  # resize image
  rgb_image = cv2.resize(rgb_image, (50, 50))

  list_image.append(rgb_image)

fig = plt.figure(figsize=(10, 10))
# define width and height
columns = 5
rows = 5

for i in range(1, columns*rows + 1):
  fig.add_subplot(rows, columns, i)
  plt.axis('on') # tắt hiển thị 2 trục
  plt.imshow(list_image[i-1])
```
